```python
import pandas as pd
import json
import os
import numpy as np
from datetime import date
```

```python
# Define the file path (update with your actual file path)
file_path = "./data/updated_accessibility_restaurant_info.json"

# Ensure the file exists
if os.path.exists(file_path):
    # Load the JSON data
    with open(file_path, 'r') as f:
        data = json.load(f)

    # Convert the JSON data into a pandas DataFrame
    df = pd.DataFrame(data)

    # Display the DataFrame
    print(df)
else:
    print(f"File not found: {file_path}")
```

```python
# Add new columns with "Unknown" as the default value
df['wheelchair_accessible_parking'] = "Unknown"
df['wheelchair_accessible_restroom'] = "Unknown"
df['wheelchair_accessible_seating'] = "Unknown"

# Display the updated DataFrame
df.head()
```

```python
# Desired column order
new_column_order = [
    "name", 
    "wheelchair_accessible_entrance", 
    "wheelchair_accessible_parking", 
    "wheelchair_accessible_restroom", 
    "wheelchair_accessible_seating",
    "latitude", 
    "longitude", 
    "place_id" 
]

# Reassign the DataFrame with the new column order
df = df[new_column_order]

# Display the DataFrame to check the new column order
df.head()
```

```python
# Define the directory and file name
directory = "./data/"

# Use today's date in the filename
file_name = f"{date.today()}_accessible_restaurant_info.json"

# Create the full file path
file_path = os.path.join(directory, file_name)

# Save the DataFrame to the JSON file with the full file path
df.to_json(file_path, orient="records", indent=4)
```

```python
import pandas as pd

# Load Table 1 (name, place_id, address) from JSON file
table1 = pd.read_json('./data/shifted_restaurants_list.json')

# Load Table 2 (name, place_id, accessibility info) from JSON file
table2 = pd.read_json('./data/updated_accessibility_restaurant_info.json')
```

```python
# Perform an inner join on 'place_id'
combined_df = pd.merge(table1, table2, on='place_id', how='inner')

# Preview the result
print(combined_df.head())
```

```python
# Define the directory and file name
directory = "./data/"

# Use today's date in the filename
file_name = f"{date.today()}_accessible_restaurant_info.json"

# Create the full file path
file_path = os.path.join(directory, file_name)

# Save the DataFrame to the JSON file with the full file path
combined_df.to_json(file_path, orient="records", indent=4)
```
